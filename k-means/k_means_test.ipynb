```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt


# using sklearn's make_blobs function for generating a test dataset
X, true_labels = make_blobs(n_samples=300, centers=3, n_features=2, cluster_std=2, random_state=42)

# Quick plot
plt.scatter(X[:, 0], X[:, 1], alpha=0.7, s=15)
plt.title("Test data")
plt.show()
```

```python
import numpy as np
from k_means_simple import k_means

labels, centers, history = k_means(X, k=3)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, alpha=0.7, s=15)
plt.scatter(centers[:, 0], centers[:, 1], marker='+', c='fuchsia', s=300)
plt.title("K-Means Test")
plt.show()

#
```
